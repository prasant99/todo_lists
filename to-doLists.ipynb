```python
#Function to let user add new tasks
def add_tasks():
    task = input("Enter the task you want to do: ")
    with open("tasks.txt", "a") as file:
        file.write(task + "\n")
        print("Task has been added.")
#Function to let users view their to-do lists
def view_tasks():
    #Handeling error for no file and no tasks
    try:
        with open("tasks.txt", "r") as file:
            tasks = file.readlines()
            if tasks:
                print("\nYour tasks: ")
                for i, task in enumerate (tasks, 1):
                    print(f"{i}. {task.strip()}")
            else:
                print("No tasks to do yet!")
    except FileNotFoundError:
        print("No tasks to do yet!")


    

#Create a menu for user
def menu():
    print("\nMenu: ")
    print("1. Add task")
    print("2. View tasks")
    print("3. Mark task done")
    print("4. Delete task")
    print("5. Exit")
#Loop runs until user exists
while True:
    menu()
    choice = input("What do you want to do?: ")
    if choice == "1":
        add_tasks()
    elif choice == "2":
        view_tasks()
    elif choice == "5":
        print("Bye!")
        break
```
